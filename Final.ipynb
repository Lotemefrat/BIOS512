```r
heart_health <- read.csv('heart_disease_health_indicators_BRFSS2015.csv')
heart_health
```

```r
colnames(heart_health)
```

```r
library(ggplot2)
```

```r
ggplot(heart_health, aes(x = as.factor(HeartDiseaseorAttack), y = BMI)) +
    geom_boxplot() +
    labs(
        title = "BMI Boxplot",
        x = "Heart Disease or Attack (0 = No, 1 = Yes)",
        y = "BMI"
    )  +
  theme_minimal()
```

```r
ggplot(heart_health, aes(x = BMI)) +
  geom_density() +
  facet_wrap(~GenHlth) +
  labs(
    title = "BMI by General Health Category",
    x = "BMI"
  ) +
  theme_minimal()
```

```r
# Calculate mean BMI for each GenHlth category
mean_bmi_table <- aggregate(BMI ~ GenHlth, data = heart_health, FUN = mean)
colnames(mean_bmi_table) <- c("General Health", "Mean BMI")
mean_bmi_table
```

My analysis focuses on BMI and health. The first visual is a box plot comparing the BMI distribution between patients who had a heart disease/attack compared to those who did not. We can see that individuals who didn't have a heart disease/attack have a lower median, first quartile, and third quartile. It appears that the maximum and mimimum BMI value is approximately equal among the groups. This suggests the heart disease/attack group has a higher BMI than the no heart disease/attack group.

The second visual compares BMI distributions between general health categories (1=Excellent, 2=Very good, 3=Good, 4=Fair, 5=Poor). By observing the density distributions we can see that as the category increases, the peak broadens towards the higher BMI values. Visual three supports this observation, demostrating the trend that as the category increases, going from excellent to poor general health, mean BMI increases, suggesting that those with poorer general health have higher BMI.
